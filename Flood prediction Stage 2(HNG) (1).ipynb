```python
#drop unneeded columns

df = df.drop(columns =['windspeedmin', 'windspeedmax', 'windgust',  'stations', 'icon', 'moonphase', 'uvindex',  'severerisk', 'solarradiation', 'solarenergy', 'cloudcover', 'feelslikemax', 'feelslikemin', 'feelslike', 'dew',  'snow', 'snowdepth'])
```

```python
# Convert the 'datetime' column to datetime format
df['datetime'] = pd.to_datetime(df['datetime'])

# Extract year and month for better analysis
df['year'] = df['datetime'].dt.year
df['month'] = df['datetime'].dt.month

# Group by year and month to get monthly total precipitation
monthly_precipitation = df.groupby(['year', 'month'])['precip'].sum().reset_index()

# Plot the monthly total precipitation
plt.figure(figsize=(14, 7))
plt.plot(monthly_precipitation['year'].astype(str) + '-' + monthly_precipitation['month'].astype(str), 
         monthly_precipitation['precip'], marker='o', linestyle='-')
plt.xlabel('Date (Year-Month)')
plt.ylabel('Total Precipitation (mm)')
plt.title('Monthly Total Precipitation in Lagos')
plt.xticks(rotation=90)
plt.grid(True)
plt.savefig('Monthly_total_precipitation_in_Lagos.png')
plt.show()
```

```python
# Seasonal decomposition to identify patterns
from statsmodels.tsa.seasonal import seasonal_decompose

# Resample the data to monthly frequency and fill missing values
monthly_precip = df.set_index('datetime')['precip'].resample('M').sum()

# Perform seasonal decomposition
decomposition = seasonal_decompose(monthly_precip, model='additive', period=12)

# Plot the decomposed components
plt.figure(figsize=(14, 10))
plt.subplot(411)
plt.plot(decomposition.observed, label='Observed')
plt.legend(loc='upper left')
plt.subplot(412)
plt.plot(decomposition.trend, label='Trend')
plt.legend(loc='upper left')
plt.subplot(413)
plt.plot(decomposition.seasonal, label='Seasonal')
plt.legend(loc='upper left')
plt.subplot(414)
plt.plot(decomposition.resid, label='Residual')
plt.legend(loc='upper left')
plt.tight_layout()
plt.show()
```

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing

# Fit the Exponential Smoothing model
model = ExponentialSmoothing(monthly_precip, seasonal='additive', seasonal_periods=12)
fitted_model = model.fit()

# Forecast the next 12 months
forecast = fitted_model.forecast(36)

# Plot the actual data and the forecast
plt.figure(figsize=(14, 7))
plt.plot(monthly_precip, label='Actual')
plt.plot(forecast, label='Forecast', linestyle='--', color='red')
plt.xlabel('Date')
plt.ylabel('Total Precipitation (mm)')
plt.title('Monthly Precipitation Forecast for Lagos')
plt.legend()
plt.grid(True)
plt.savefig('monthly_precipitation_forecast_for_lagos.png')
plt.show()

print("Forecasted Precipitation:\n", forecast)
```

```python

# Split the data into training and testing sets
train_size = int(len(monthly_precip) * 0.8)
train, test = monthly_precip[:train_size], monthly_precip[train_size:]
```

```python

# Define the flood threshold (e.g., mean + 2 standard deviations)
threshold = monthly_precip.mean() +  monthly_precip.std()
print("Flood Threshold:", threshold)

# Identify potential flood months
potential_floods = forecast[forecast > threshold]
print("Potential flood months:", potential_floods)
```

```python

# Identify historical extreme values
historical_extreme_values = monthly_precip[monthly_precip > threshold]
print("Historical Extreme Values:\n", historical_extreme_values)

# Print the dates with extreme values
extreme_dates = historical_extreme_values.index
print("Dates with Extreme Precipitation Values (Potential Flooding):")
for date in extreme_dates:
    print(date.strftime('%Y-%m'))
```

```python
import seaborn as sns

df['datetime'] = pd.to_datetime(df['datetime'])
df.set_index('datetime', inplace=True)



monthly_precip = df['precip'].resample('M').sum()
# Pivot the data to create a heatmap
monthly_precip_pivot = monthly_precip.reset_index()
monthly_precip_pivot['Year'] = monthly_precip_pivot['datetime'].dt.year
monthly_precip_pivot['Month'] = monthly_precip_pivot['datetime'].dt.strftime('%b')
pivot_table = monthly_precip_pivot.pivot(index='Month', columns='Year', values='precip')

# Sort the months in the correct order
months_order = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
pivot_table = pivot_table.reindex(months_order)

# Create the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(pivot_table, cmap='Blues', annot=True, fmt=".1f", linewidths=.5)
plt.title('Seasonal Rainfall Heatmap')
plt.xlabel('Year')
plt.ylabel('Month')
plt.savefig('seasonal_rainfall_heatmap.png')
plt.show()
```

```python
# Threshold value
threshold = monthly_precip.mean() +  monthly_precip.std()

# Flood Threshold Visualization
plt.figure(figsize=(14, 7))
plt.plot(monthly_precip.index, monthly_precip, label='Monthly Precipitation')
plt.axhline(y=threshold, color='r', linestyle='--', label=f'Threshold: {threshold:.2f} mm')
plt.xlabel('Date')
plt.ylabel('Total Precipitation (mm)')
plt.title('Monthly Precipitation with Flood Threshold')
plt.legend()
plt.grid(True)
plt.savefig('flood_threshold_plot.png')
plt.show()
```

```python
# Yearly Total Precipitation
monthly_precip = df['precip'].resample('M').sum()
yearly_precip = monthly_precip.resample('Y').sum()
yearly_precip.index = yearly_precip.index.year
plt.figure(figsize=(14, 7))
yearly_precip.plot(kind='bar', color='skyblue')
plt.xlabel('Year')
plt.ylabel('Total Precipitation (mm)')
plt.title('Yearly Total Precipitation in Lagos')
plt.grid(True)
# Format the x-axis to show only the year
#plt.gca().xaxis.set_major_formatter(plt.matplotlib.dates.DateFormatter('%Y'))

plt.savefig('yearly_total_precipitation_plot.png')
plt.show()
```

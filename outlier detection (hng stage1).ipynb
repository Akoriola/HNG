### Neighbour Identification

```python
from geopy.distance import geodesic
from tqdm import tqdm  # corrected import statement

# Function to find neighboring polling units within a 1 km radius
def find_neighbours(df, radius=1):
    neighbours = []
    for index, row in tqdm(df.iterrows(), total=df.shape[0], desc='Processing units'):
        unit_location = (row['Latitude'], row['Longitude'])
        neighbour_units = df[df.apply(lambda x: geodesic(unit_location, (x['Latitude'], x['Longitude'])).km < radius, axis=1) & (df.index != index)]
        neighbours.append(neighbour_units.index.tolist())
        #neighbours.append(neighbour_units['PU-Code'].tolist())  # Replace 'PU-Code' with the actual column name for PU codes
    return neighbours

# Identify neighbors
df['Neighbours'] = find_neighbours(df)
df.head()
```

```python
#rename Neighbours to Neighbouring units
df.rename(columns={'Neighbours': 'Neighbouring units'}, inplace=True)

# to check
df.head(10)
```

### Outlier Score Calculation

```python
def calculate_outlier_scores(df):
    parties = ['APC', 'LP', 'PDP', 'NNPP']  # Update this list with actual party columns in your dataset
    
    # Initialize columns for outlier scores
    for party in parties:
        df[f'{party}_Outlier_Score'] = 0
    
    for index, row in df.iterrows():
        neighbours = df.loc[row['Neighbouring units']]
        for party in parties:
            if not neighbours.empty:
                score = np.abs(row[party] - neighbours[party].mean())
                df.at[index, f'{party}_Outlier_Score'] = score
    
    return df

# Calculate outlier scores and update the DataFrame
df = calculate_outlier_scores(df)

# Display the first few rows of the updated DataFrame
df.head()
```

### Sort and Report

```python
sorted_apc = df.sort_values(by='APC_Outlier_Score', ascending=False).head(3)
sorted_lp = df.sort_values(by='LP_Outlier_Score', ascending=False).head(3)
sorted_pdp = df.sort_values(by='PDP_Outlier_Score', ascending=False).head(3)
sorted_nnpp = df.sort_values(by='NNPP_Outlier_Score', ascending=False).head(3)


# Display the top 3 outliers for each party
print("Top 3 APC Outliers:")
print(sorted_apc)
print("\nTop 3 LP Outliers:")
print(sorted_lp)
print("\nTop 3 PDP Outliers:")
print(sorted_pdp)
print("\nTop 3 NNPP Outliers:")
print(sorted_nnpp)
```

### Excel file with the sorted outliers scores

```python

# Function to create a DataFrame for each party with the specified columns
def create_outlier_df(party, df):
    columns = ['PU-Code', 'PU-Name', 'Party', 'Outlier_Score', 'Neighbouring units']
    outlier_df = df[['PU-Code', 'PU-Name', f'{party}_Outlier_Score', 'Neighbouring units']].copy()
    outlier_df = outlier_df[outlier_df[f'{party}_Outlier_Score'] > 0]  # Only include rows with non-zero outlier scores
    outlier_df['Party'] = party
    outlier_df = outlier_df.rename(columns={f'{party}_Outlier_Score': 'Outlier_Score'})
    return outlier_df[columns]

# Create DataFrames for each party
apc_outliers = create_outlier_df('APC', df)
lp_outliers = create_outlier_df('LP', df)
pdp_outliers = create_outlier_df('PDP', df)
nnpp_outliers = create_outlier_df('NNPP', df)

# Sort each DataFrame by 'Outlier_Score'
sorted_apc = apc_outliers.sort_values(by='Outlier_Score', ascending=False)
sorted_lp = lp_outliers.sort_values(by='Outlier_Score', ascending=False)
sorted_pdp = pdp_outliers.sort_values(by='Outlier_Score', ascending=False)
sorted_nnpp = nnpp_outliers.sort_values(by='Outlier_Score', ascending=False)

# Save each sorted DataFrame to an Excel file
with pd.ExcelWriter(r'C:\Users\Olasunkanmi Dara\Downloads\sorted_outlier_scores.xlsx') as writer:
    sorted_apc.to_excel(writer, sheet_name='APC_Outliers', index=False)
    sorted_lp.to_excel(writer, sheet_name='LP_Outliers', index=False)
    sorted_pdp.to_excel(writer, sheet_name='PDP_Outliers', index=False)
    sorted_nnpp.to_excel(writer, sheet_name='NNPP_Outliers', index=False)



```
